## 고민 
- 주택소유상태(4종류), 대출목적(12종류)을 label encoding해야 할까 아니면 one hot encoding을 해야 할까?
- 대출기간은 두 종류 뿐이니 label encoding이 나을까?
- 근로기간은 다양하고 오탈자도 있다. Unknown의 경우는 무엇을 넣는 것이 가장 합리적일까?

## 시도
- 근로 기간의 경우는 오탈자를 제외하고 label encoding하는게 좋아보임.
- 나머지는 원핫인코딩이 낫지 않을까
- label로 하다가 숫자가 커짐이 의미 있는 것으로 해석될 가능성이 있어 보임. (특히 회귀에서)

```python
from sklearn.preprocessing import OneHotEncoder

oh = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoded_data = oh.fit_transform(df[['주택소유상태', '대출목적', '대출기간']])
encoded_data
```

```python
# 새로운 데이터프레임에 인코딩된 데이터 추가
df_encoded = pd.concat([df, pd.DataFrame(encoded_data, columns=oh.get_feature_names_out(['주택소유상태', '대출목적', '대출기간']))], axis=1)
df_encoded
```

```python
df_encoded.drop(columns=['대출목적', '주택소유상태', '대출기간'], inplace= True)
df_encoded
```

```python
# 일단 근로기간도 제외를 하자
df_encoded.drop(columns = ['근로기간'], inplace=True)
df_encoded
```

```python
# 대출등급을 숫자로 매핑하는 사전(dictionary) 생성
grade_mapping = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7}

# 대출등급 컬럼 변환
df_encoded['대출등급'] = df_encoded['대출등급'].map(grade_mapping)
```

```python
train = df_encoded.drop(['ID','대출등급'], axis=1)
target = df_encoded['대출등급']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train, target, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import f1_score
def evaluate_f1(y_test, y_predict):
    F1 = f1_score(y_test, y_predict, average='weighted')
    print(f'F1: {F1:.3F}')
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr = RandomForestClassifier()
```

```python
rfr.fit(x_train, y_train)
```

```python
y_predict = rfr.predict(x_test)
evaluate_f1(y_test, y_predict)
```

```python
# 숫자를 알파벳 등급으로 변환하는 함수
def map_to_grade(number):
    reverse_grade_mapping = {1: 'A', 2: 'B', 3: 'C', 4: 'D', 5: 'E', 6: 'F', 7: 'G'}
    return reverse_grade_mapping[number]

# 숫자로 예측된 등급을 알파벳 등급으로 변환
predicted_grades = [map_to_grade(prediction) for prediction in test_1]

# 예측된 등급과 ID를 포함한 DataFrame 생성
result_df = pd.DataFrame({'ID': test_df['ID'], '대출등급': predicted_grades})

# CSV 파일로 저장
result_df.to_csv('sub_1.csv', index=False)
```

## Unknown 전처리 하기

```python
# 규칙에 따라 문자열 수정
df['근로기간'] = df['근로기간'].str.replace(' years', '')  # years 삭제
df['근로기간'] = df['근로기간'].str.replace('+years', '')
df['근로기간'] = df['근로기간'].str.replace('10+years', '10')
df['근로기간'] = df['근로기간'].str.replace('10+', '10')
df['근로기간'] = df['근로기간'].str.replace(' year', '')   # year 삭제
df['근로기간'] = df['근로기간'].str.replace('< 1', '0.9')
df['근로기간'] = df['근로기간'].str.replace('<1', '0.9')

# 결과 확인
df['근로기간'].value_counts()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
# Unknown 항목을 예측하기 위해 데이터프레임을 훈련 데이터와 테스트 데이터로 나눔
train_data = df[df['근로기간'] != 'Unknown']
test_data = df[df['근로기간'] == 'Unknown']

# 훈련 데이터에서 근로기간 컬럼을 제외한 나머지 컬럼들을 X로, 근로기간을 y로 설정
X_train = train_data.drop(columns=['ID', '근로기간'], axis=1)
y_train = train_data['근로기간'].astype(float)

# Random Forest 모델 훈련
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)

# 테스트 데이터에서 근로기간을 예측
X_test = test_data.drop(columns=['ID','근로기간'], axis=1)
predicted_values = rfr.predict(X_test)

# 예측값을 데이터프레임에 채워 넣음
df.loc[df['근로기간'] == 'Unknown', '근로기간'] = predicted_values

df
```
