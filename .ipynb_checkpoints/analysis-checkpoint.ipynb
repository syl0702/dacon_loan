- 근로 기간의 경우는 오탈자를 제외하고 label encoding하는게 좋아보임.
- 나머지는 원핫인코딩이 낫지 않을까
- label로 하다가 숫자가 커짐이 의미 있는 것으로 해석될 가능성이 있어 보임. (특히 회귀에서)
