## 고민 
- 주택소유상태(4종류), 대출목적(12종류)을 label encoding해야 할까 아니면 one hot encoding을 해야 할까?
- 대출기간은 두 종류 뿐이니 label encoding이 나을까?
- 근로기간은 다양하고 오탈자도 있다. Unknown의 경우는 무엇을 넣는 것이 가장 합리적일까?

## 시도
- 근로 기간의 경우는 오탈자를 제외하고 label encoding하는게 좋아보임.
- 나머지는 원핫인코딩이 낫지 않을까
- label로 하다가 숫자가 커짐이 의미 있는 것으로 해석될 가능성이 있어 보임. (특히 회귀에서)

```python
from sklearn.preprocessing import OneHotEncoder

oh = OneHotEncoder(sparse=False)
encoded_data = oh.fit_transform(df[['주택소유상태', '대출목적', '대출기간']])
encoded_data
```

```python
# 새로운 데이터프레임에 인코딩된 데이터 추가
df_encoded = pd.concat([df, pd.DataFrame(encoded_data, columns=oh.get_feature_names_out(['주택소유상태', '대출목적', '대출기간']))], axis=1)
df_encoded
```

```python
df_encoded.drop(columns=['대출목적', '주택소유상태', '대출기간'], inplace= True)
df_encoded
```

```python
# 일단 근로기간도 제외를 하자
df_encoded.drop(columns = ['근로기간'], inplace=True)
df_encoded
```

```python
# 대출 등급 원핫인코딩
oh = OneHotEncoder(sparse=False)
encoded_data = oh.fit_transform(df[['대출등급']])
df_encoded = pd.concat([df_encoded, pd.DataFrame(encoded_data, columns=oh.get_feature_names_out(['대출등급']))], axis=1)
df_encoded
```

```python
train = df_encoded.drop(columns=['ID', '대출등급', '대출등급_A', '대출등급_B', '대출등급_C', '대출등급_D', '대출등급_E', '대출등급_F', '대출등급_G'])
target = df_encoded[['대출등급_A', '대출등급_B', '대출등급_C', '대출등급_D', '대출등급_E', '대출등급_F', '대출등급_G']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train, target, test_size=0.3)
```

```python
from sklearn.metrics import f1_score
def evaluate_f1(y_test, y_predict):
    F1 = f1_score(y_test, y_predict)
    print(f'F1: {F1:.3F}')
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
```

```python
rfr.fit(x_train, y_train)
```

```python
y_predict = rfr.predict(x_test)
evaluate_f1(y_test, y_predict)
```

```python
from sklearn.metrics import accuracy_score

from sklearn.metrics import accuracy_score

# 실제 레이블
y_true = df_encoded['대출등급']

# 모델의 예측 결과
y_pred = rfr.predict(train)

# 정확도 계산
accuracy = accuracy_score(y_true, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```
